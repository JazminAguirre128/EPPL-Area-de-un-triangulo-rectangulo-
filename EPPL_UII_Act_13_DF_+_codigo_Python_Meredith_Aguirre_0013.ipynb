```python
# Calcular area de un circulo
# variables B, h, A
# formula A = (B * h)/2
# Captura el valor de A
B = float(input("Ingrese el valor de la base: "))
h = float(input("Ingrese el valor de la altura: "))
A = (B * h)/2
print("El area del triangulo rectangulo es: ", A)
print("Meredith Aguirre NL = 2")
```
